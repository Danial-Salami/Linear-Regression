```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
X = iris.data[:,[1,3]]
y = iris.target
```

```python
model = SVC(kernel='linear',C=5,gamma='auto')
```

```python
model.fit(X,y)
```

```python
plot_decision_regions( X,y,clf=model,legend=1);
```

```python
from sklearn.datasets import make_moons
X1, y1 = make_moons(n_samples=100, random_state=42)
```

```python
model1 = SVC(kernel='rbf', gamma='auto', C=2)

model1.fit(X1, y1);
plot_decision_regions(X1, y1, clf = model1 , legend=2 );
```

```python
from sklearn.datasets import make_circles

X2, y2 = make_circles(n_samples=1000, random_state=42, noise=0.15, factor=0.25)

model2 = SVC(kernel='rbf', gamma='auto', C=2)
model2.fit(X2, y2);
plot_decision_regions(X2, y2, clf = model2 , legend=2 );
```

```python
#xx, yy = np.meshgrid(np.linspace(-2, 2, 40),np.linspace(-2, 2, 40))

rng = np.random.RandomState(1)
X3 = rng.randn(300, 2)

y3 = np.array(np.logical_xor(X3[:, 0] > 0, X3[:, 1] > 0), dtype=int)


model3 = SVC(kernel='rbf', gamma='auto', C=0.5)

model3.fit(X3, y3)

plot_decision_regions(X3, y3, clf = model3 , legend=2 );
```
